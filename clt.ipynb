```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Generating the 1000 samples from chi square distribution

```python
k=3
r_chi2=sts.chi2(k)
sample = r_chi2.rvs(1000)
```

[__*Chi square distribution*:__](https://en.wikipedia.org/wiki/Chi-square_distribution)
$$
X_{2}^k
$$
$$
k=3  
$$
$$
M[x]=k=3
$$
$$
D[x]=2k=6
$$

__*Histogram and distribution density*__

```python
x = np.linspace(0,20,100)
pdf = r_chi2.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.hist(sample, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

### Estimation of the distribution of the sample mean of our random variable

__n=5__

```python
n=5
sample_n_5=[]
for i in range(1,1001):
    sample_n_5.append(np.mean(np.random.choice(sample,n)))
```

```python
x = np.linspace(0,20,100)
pdf = sts.norm(k, ((2*k)/n)**(0.5)).pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.hist(sample_n_5, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

$$
m[x]=M[x]=3   
$$
$$
d[x]={D[x]\over 5}=1.2
$$
(*m*[x] and *d*[x] are mean and variance of new distibution)

__Conclusion:__ If we increase the sample size of the sample mean of a random variable, we will see the distibution similar to normal with mean and variance equal to M[x] and D[x]/n
